#Data Collection + Cleaning: California Department of Education School Locations and Demographic Data

```python
import pandas as pd
geo_2020 = pd.read_csv("/content/SchoolSites2021_-5298042096866201909.csv")
geo_2021 = pd.read_csv("/content/SchoolSites2122_-5936143432603304851.csv")
geo_2022 = pd.read_csv("/content/SchoolSites2223_-3988485820510116362.csv")
geo_2023 = pd.read_csv("/content/SchoolSites2324_4153587227043982744.csv")
df_merged_filtered = pd.read_csv("/content/merged_cleaned_data(in).csv")
geo_2020
```

```python
def process_for_year(year, geo_df, df_merged_filtered):
    df = df_merged_filtered[df_merged_filtered["School Year"] == year]
    dup_rows = df.duplicated(subset = ["School_Code"], keep="first")
    df = df[~dup_rows]

    common_school_code = geo_df["School Code"].isin(df["School_Code"])
    filtered_geo_df = geo_df[common_school_code]

    merged_df = pd.merge(df, filtered_geo_df, left_on="School_Code", right_on="School Code")
    result_df = merged_df[['Percentage Standard Exceeded', 'Percentage Standard Met','Percentage Standard Met and Above', 'Percentage Standard Nearly Met','Percentage Standard Not Met']]
    result_df = result_df.replace('*', float("nan"))
    result_df = result_df.apply(pd.to_numeric, errors='coerce')
    result_df['Socioeconomically Disadvantaged'] = merged_df['Socioeconomically Disadvantaged (%)'].replace('*', float("nan"))
    result_df["Socioeconomically Disadvantaged"] = pd.to_numeric(result_df['Socioeconomically Disadvantaged'], errors='coerce')
    result_df = result_df.dropna()

    return result_df, merged_df
```

Merged datasets based off the school years (2020-2024)

```python
geo_dfs = {
    "2020-2021": geo_2020,
    "2021-2022": geo_2021,
    "2022-2023": geo_2022,
    "2023-2024": geo_2023
}
result_dfs = []
merged_dfs = []
for year, geo_df in geo_dfs.items():
    result_df, merged_df = process_for_year(year, geo_df, df_merged_filtered)
    result_dfs.append((year, result_df))
    merged_dfs.append((year, merged_df))
```

Shows the the merged data from the school year 2020-2021

```python
result_df = result_dfs[0][1]
merged_df = merged_dfs[0][1]
```

#Visualizing the Correlation data

```python
import plotly.express as px
all_years_data = []
for year, result_df in result_dfs:
    # Add a 'Year' column to the current year's data
    result_df['Year'] = year

    # Append the data to the list
    all_years_data.append(result_df)

all_years_df = pd.concat(all_years_data, ignore_index=True)
cols_to_plot = ['Percentage Standard Met and Above', 'Percentage Standard Nearly Met',
                'Percentage Standard Not Met']
correlation_data = []
for year in all_years_df['Year'].unique():
    for col in cols_to_plot:
        corr_value = all_years_df[all_years_df['Year'] == year][col].corr(
            all_years_df[all_years_df['Year'] == year]['Socioeconomically Disadvantaged']
        )
        correlation_data.append({'Year': year, 'Category': col, 'Correlation': corr_value})

# Create a DataFrame from the correlation data
correlation_df = pd.DataFrame(correlation_data)

fig = px.line(
    correlation_df,
    x='Year',
    y='Correlation',
    color='Category',
    title='Correlation Between Percentage Scores and Socioeconomically Disadvantaged Over Time'
)

# Update layout
fig.update_layout(xaxis_title='Year', yaxis_title='Correlation')

# Show the graph
fig.show()
```

# Interpretation:
Graph provides insight into the relationship between student perfomance on standardized tests and socioeconomic disadvantage over time.

For the "Percentage Standard Met and Above" and "Socioeconomically Disadvantaged", we can see that there is a negative correlation, means that as the percentage of socioeconomically disadvantaged students increases, the percentage of students meeting or exceeding the standards tends to decrease. The line is relative close to zero, suggesting a moderate negative correlation that's not extremely strong. The line slopes down betwen 2020-2021 and 2021-2022, showing that the negative correlation is becoming stronger. From 2021-2022 and onwards, can see that its a slight upward slope, slightly moving towards a positive correlation but not by much.

For the "Percentage Standard Nearly Met" and "Socioeconomically Disadvantaged", we can see that there is a positive correlation. The "Percentage Standard Nearly Met" represents the proportion of students who scored just below the proficiency level required to "meet" the standard. They have demonstrated good understanding of the concepts but doesn't have full mastery. This positive correlation suggests that schools with a higher percentage of socioeconomically disadvantaged students tend to have a higher percentage of students who nearly met the standard. Since its around 0.35 (the correlation), it is seen as a weaker to moderate correlation.


For the "Percentage Standard Not Met", this metric represnets the proportion of students who don't demonstrate proficiency in the tested concepts and fell below the minimum level of "nearly met". This is a postive correlation, suggesting that schools with a higher percentage of students of socioeconomically disadvantaged students tend to have a higher percentage of students who don't meet the standard.

#Visualizing the Scatter plot Data

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# List of columns to plot
cols_to_plot = ['Percentage Standard Met and Above', 'Percentage Standard Nearly Met',
                'Percentage Standard Not Met']

# Determine number of rows and columns
rows = 3  # 3 rows (one with a single plot)
cols = 2  # 2 columns

# Create a dictionary to store figures for each year
figs = {}

# Iterate through each year's data
for year, result_df in result_dfs:
    # Create subplots for the current year
    fig = make_subplots(rows=rows, cols=cols, subplot_titles=cols_to_plot)

    # Add traces for each column
    for i, col in enumerate(cols_to_plot):
        row_num = i // cols + 1  # Calculate row number
        col_num = i % cols + 1   # Calculate column number

        # Extract x and y data
        x_data = result_df[col]
        y_data = result_df['Socioeconomically Disadvantaged']

        # Scatter plot
        fig.add_trace(
            go.Scatter(x=x_data, y=y_data, mode='markers', name=f'{col} Scatter'),
            row=row_num, col=col_num
        )

        # Line of best fit
        if len(x_data) > 1:
            slope, intercept = np.polyfit(x_data, y_data, 1)
            trendline = slope * np.array(x_data) + intercept
            fig.add_trace(
                go.Scatter(x=x_data, y=trendline, mode='lines',
                           line=dict(color='black'), name=f'{col} Trendline'),
                row=row_num, col=col_num
            )

    # Update layout
    fig.update_layout(height=900, title_text=f"{year}: Percentage Scores vs Socioeconomically Disadvantaged")

    # Store the figure in the dictionary
    figs[year] = fig

# Display the figures for each year
for year, fig in figs.items():
    fig.show()
```
